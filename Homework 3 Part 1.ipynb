```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
networkData = pd.read_csv('Social_Network_Ads.csv')
print(networkData.columns)
print(networkData.dtypes)
print(networkData.shape)
```

```python
inputData = networkData.drop(['Purchased','User ID'],axis=1)
inputData1 = pd.get_dummies(inputData)
print(inputData1)
norm_data = preprocessing.normalize(inputData1, axis=0)
norm_df = pd.DataFrame(norm_data,columns=[inputData1.columns])
print(norm_df)
```

```python
Y = networkData.drop(['Age','User ID','EstimatedSalary','Gender'],axis=1)
X_train, X_test, y_train, y_test = train_test_split(norm_df, Y, train_size = 0.7, random_state = 0)
```

```python
clf =  RandomForestClassifier(oob_score=True,bootstrap=True)
errors = []
estimators = []
for i in range(5,151):
    clf.set_params(n_estimators=i)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    estimators.append(i)
    oob_error = 1 - clf.oob_score_
    errors.append(oob_error)
```

```python
plt.figure(figsize=(15,6))
plt.plot(estimators, errors)
plt.xlabel("n-estimators")  # add X-axis label
plt.ylabel("Oob errors")  # add Y-axis label
plt.title("oob error vs number of trees.") 
plt.show(estimators,errors)
```

```python
for i in range(0,146):
    if errors[i] < 0.115:
        print(str(estimators[i]) + ": " + str(errors[i]))
```
