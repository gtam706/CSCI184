```python
from sklearn import preprocessing
import pandas as pd
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
```

```python
data = pd.read_csv('Wholesale customers data.csv')
normalizedData = preprocessing.normalize(data)
df = pd.DataFrame(normalizedData,columns=[data.columns])
print(data)
print(df)
```

```python
plt.figure(figsize=(100, 20))
plt.title('Dendrogram of Customers dataset')
dend = shc.dendrogram(shc.linkage(normalizedData, method='complete'))
```

```python
plt.figure(figsize=(100, 20))
plt.title('Dendrogram of Customers dataset')
dend = shc.dendrogram(shc.linkage(normalizedData, method='single'))
```

```python
plt.figure(figsize=(100, 20))
plt.title('Dendrogram of Customers dataset')
dend = shc.dendrogram(shc.linkage(normalizedData, method='average'))
```

```python
predictedResults = {}
for n in range(2,6):
    cluster = AgglomerativeClustering(n_clusters=n,linkage='complete')
    predictedResults[str(n)] = cluster.fit_predict(normalizedData)
print(predictedResults)
```
