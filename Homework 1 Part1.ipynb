```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
import pandas as pd
data = pd.read_csv('train.csv')
print(data)
print(data.shape)
```

```python
from pandas.api.types import is_string_dtype
data = data.drop('Id',axis=1)
data = data.dropna(axis=1)
for col in data:
    if is_string_dtype(data[col]) and data[col].nunique() >= 10:
        data = data.drop(col,axis=1)
display(data)
print(data.shape)
```

```python
from sklearn.preprocessing import OneHotEncoder
x = data.drop('SalePrice',axis=1)
xa = x
xb = pd.get_dummies(x)
y = data['SalePrice']
for col in data:
    if is_string_dtype(data[col]):
        xa = xa.drop(col,axis=1)
display(xa)
display(xb)
```

```python
lm = LinearRegression()
model = lm.fit(xa,y)
pred_y = model.predict(xa)
print(f'absolute mean error for only numerical columns= {mean_absolute_error(y, pred_y)}')

lm = LinearRegression()
model = lm.fit(xb,y)
pred_y = model.predict(xb)
print(f'absolute mean error for one-hot encoding columns= {mean_absolute_error(y, pred_y)}')


```
